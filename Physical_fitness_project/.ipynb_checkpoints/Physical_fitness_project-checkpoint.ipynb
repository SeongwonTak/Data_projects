```python
def is_number(num):
    try:
        float(num)
        return True #num을 float으로 변환할 수 있는 경우
    except ValueError: #num을 float으로 변환할 수 없는 경우
        return False
```

```python
exercise = non_handi_data.columns[8:]
for col in exercise:
    non_numeric = []
    for i in range(len(non_handi_data)):
        if not is_number(str(non_handi_data[col][i])):
            non_numeric.append((non_handi_data[col][i]))
    print(col)
    print(set(non_numeric))
```

결측치 처리를 다음과 같이 할 수 있을 것이다.  
(1) 연령대를 계산후, 연령대별 평균을 낸다.

```python
def get_gen(age):
    return str(age//10 * 10) + 's'
```

```python
non_handi_data['gen'] = non_handi_data.age.apply(lambda x : get_gen(x))
```

연령대별로 평균을 구하는 것이 합리적일 것이다.
따라서, 다음과 같이 하려고 한다.

(1) 연령대별로 기록을 본다.  
(2) 기록이 없는 경우는 연령대 평균을 기준으로 한다.  
(3) 만일 연령대별 평균 기록조차 없을 경우 가장 근처의 어린 나이대를 반영  
예시> 20대, 30대, 50대 기록만 있으면 40대는 30대 기록의 평균 반영  

```python
non_handi_data['stretch_upper'].groupby(non_handi_data.gen).mean()
```

```python
non_handi_data['shuttle_run'].groupby(non_handi_data.gen).mean()
```

```python
non_handi_data['long_jump'].groupby(non_handi_data.gen).mean()
```

70대~90대는 셔틀런, 제자리멀리뛰기 기록이 없다. (사실 그럴만도 하다...)

여기서는 다음과 같이 진행하려고 한다.

```python
gen_stretch = non_handi_data['stretch_upper'].groupby(non_handi_data.gen).mean()
gen_shuttle = non_handi_data['shuttle_run'].groupby(non_handi_data.gen).mean()
gen_longjump = non_handi_data['long_jump'].groupby(non_handi_data.gen).mean()
```

```python
for i in range(len(non_handi_data)):
    cur_gen = non_handi_data.gen[i]
    if np.isnan(non_handi_data.stretch_upper[i]):
        non_handi_data.stretch_upper[i] = gen_stretch[cur_gen]
    if np.isnan(non_handi_data.shuttle_run[i]):
        if cur_gen in ['60s', '70s', '80s', '90s']:
            non_handi_data.shuttle_run[i] = 16.440553
        else:
            non_handi_data.shuttle_run[i] = gen_shuttle[cur_gen]
    if np.isnan(non_handi_data.long_jump[i]):
        if cur_gen in ['60s', '70s', '80s', '90s']:
            non_handi_data.long_jump[i] = 137.737844
        else:
            non_handi_data.long_jump[i] = gen_longjump[cur_gen]
```

### 가설 검정 - 신체 관련

#### 가설 1. 고혈압일수록 체지방량이 더 많을 것이다.

비만과 고혈압의 관계는 매우 심하다고 알고 있고, BMI보다는 체지방이 실질적인 비만과 관련된 수치로 지정된다.
이를 바탕으로 BMI와 고혈압의 관계가 정말 유의미하게 존재하는지 확인해보자.

우선 시각화를 통해 분포를 확인해보자.

```python
sns.histplot(x = 'fat_rate', hue = 'pressure_div', data = non_handi_data,
             bins = 10, alpha = 0.7)
```

이런..! 분포가 너무 겹쳐서 고혈압, 저혈압, 정상 구분의 확인이 어렵다.

다만 겹친 분포만 봐서는 정규분포를 띄는 것으로 보인다. 이제 각각의 값을 평균내보자.

```python
non_handi_data['fat_rate'].groupby(non_handi_data.pressure_div).mean()
```

정상과 저혈압, 고혈압의 큰 차이가 보이지 않는 것 같다.

```python
non_handi_data['fat_rate'].groupby(non_handi_data.pressure_div).std()
```

분산으로서도 수치적으로는 큰 차이를 보이지 않는다. 다만 통계검증을 통해 조금 더 확실하게 이게 맞는지를 확인해보자.

#### 통계 검증

```python
import scipy.stats as stats
```

```python
stats.shapiro(non_handi_data.loc[non_handi_data.pressure_div == '고혈압', 'fat_rate'])
```

```python
stats.shapiro(non_handi_data.loc[non_handi_data.pressure_div == '저혈압', 'fat_rate'])
```

```python
stats.shapiro(non_handi_data.loc[non_handi_data.pressure_div == '정상', 'fat_rate'])
```

저혈압인 경우에 대해서 비교는 조금 어려울 것 같고, 데이터 개수도 적기 때문에 넘어간다.

또한 고혈압인 경우와 정상인 경우 유의수준 1% 하에서도 정규분포가 아니라고 하므로 비모수 검정법을 사용해야 한다.

```python
mid_data = non_handi_data.loc[non_handi_data.pressure_div == '정상', 'fat_rate']
high_data = non_handi_data.loc[non_handi_data.pressure_div == '고혈압', 'fat_rate']
stats.levene(mid_data, high_data)
```

이번에는 유의수준 5% 하에서도 귀무가설이 채택되며, 등분산이라고 할 수 있다.!

```python
# 정규분포가 아니므로 mann-witneyu test를 사용해야 한다.
print(stats.mannwhitneyu(mid_data, high_data))
```

유의수준 1% 하에서도 충분히 귀무가설이 기각되어버렸다. 즉, 고혈압과 정상 사이에 충분한 체지방 차이가 있다는것이다.

평균치에서 정상보다는 고혈압 집단에서 체지방이 더 많았으므로 체지방은 고혈압을 판단하는 하나의 지표가 될 수 있을 것이다.

#### 가설검정 2 : 나이가 많으면 고혈압에 잘 걸릴 것이다.

일단 crosstab을 통해 확인해보자

```python
cross = pd.crosstab(non_handi_data.pressure_div, non_handi_data.gen)
cross
```

수치적으로만 보았을 경우 정상환자 대비 고혈압 환자의 비중이 더 높다는 것을 알 수 있다.

우선, 수치적으로 걸린 비율을 확인해보자.

```python
cols = cross.columns
for col in cols:
    print('%s : %f' %(col, cross[col]['고혈압']/sum(cross[col])))
```

다음과 같이 연령대가 높아질수록 고혈압에 해당하는 사람의 수가 늘어나고 있음을 알 수 있다.
