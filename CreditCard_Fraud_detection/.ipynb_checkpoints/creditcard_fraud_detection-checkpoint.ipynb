즉, 확실한 고금액 결제를 제외하면 대부분 fraud는 조금 더 높은 가격에서 발생했음을 알 수 있다. 그러나 분포상에는 저금액대에서도 발생했음을 알 수 있는데, 이를 구체적으로 확인하기 위해 업종별로/fraud여부별로 amt의 분포를 확인하고자 한다.

해석/ 업종별로 다시 나눠서 봤을 때 위의 경향이 더 뚜렷하게 나타났다. 고금액의 경우는 오히려 정상적인 결제인 경우다.
그러나, 해당 일부 고금액을 제외하고는 전체적으로 non-fraud 케이스에 비해 fraud 케이스가 전체적인 결제 금액이 더 높다는 것을 확인했다.

즉, 업종별로 이상치의 케이스에 대해서는 Nonfraud일 가능성이 우선적으로 생김을 확인할 수 있다.  
즉 이상치가 아닌 경우는 우선적으로는 nonfraud로 분류하면 될 것이다.  
허나 특이한 점은 일부 업종에 대해서는 fraud값이 거의 0에 가까운 지점에서 나타났는데, 이에 대해서는 다른 정보와 결합하여 특성을 파악하고자 한다.

카드 가입자는 여성보다는 남성이 조금 더 많이 분포했음을 알 수 있다.

```python
# 연령은 간단하게, 2021-year+1형태로 한국식 나이 방식으로 계산하자.
def get_age(day):
    year = int(day[:4])
    return 2022-year
```

```python
df['age'] = df.dob.apply(lambda s : get_age(s))
```

```python
df.age.describe()
```

```python
def get_gen(age):
    gen_code = str(age//10*10)
    return gen_code + 's'
```

```python
df['gen'] = df.age.apply(lambda i : get_gen(i))
```

```python
df[['dob', 'age', 'gen']]
```

다음과 같이 연령대가 잘 나눠져 있음을 확인했다. 연령대별로 성별 분포를 살펴보려고 한다.

```python
sns.countplot(x = 'gen', data = df, hue = 'gender', order = sorted(df.gen.unique()))
```

10대, 20대, 40대를 제외하고 남성이 더 많으며 특히 남성 30대의 비중이 압도적으로 높음을 확인해 볼 수 있다.   

이제 연령대별로 사기를 당한 비율이 높은 연령대가 있는지를 확인해보고자 한다.

```python
fraud_cnt_gen = (df.loc[df.is_fraud == 1]).gen.value_counts()
total_cnt_gen = df.gen.value_counts()

gen = total_cnt_gen.index
rate = []
for g in gen:
    fraud_rate = fraud_cnt_gen[g] / total_cnt_gen[g] * 100
    rate.append(fraud_rate)
```

```python
plt.figure(figsize = (16, 6))
sns.barplot(x = gen, y = rate, order = sorted(gen))
plt.xlabel('gen')
plt.ylabel('fraud_ratio')
plt.show()
```

이제 명의자 주소와 상점 주소의 거리가 fraud에 얼마나 영향을 미칠지를 확인해보고자 한다.

```python
# (lat, long)과 (merch_lat, merch_long)의 거리를 계산한다.
# 위도/경도에 대한 정보지만, 좌표의 역할을 할 수 있다고 가정 Euclidean Distance를 활용한다.
x = df.merch_lat - df.lat
y = df.merch_long - df.long

df['distance'] = np.sqrt(x ** 2 + y ** 2)
```

```python
plt.figure(figsize = (20, 6))
plt.subplot(121)
sns.histplot(x = 'distance', data = df, bins = 20)
plt.title('num of holders by distance')

plt.subplot(122)
sns.histplot(x = 'distance', data = df.loc[df.is_fraud == 1], bins = 20)
plt.title('num of holders by distance(fraud only)')

plt.show()
```

분포를 고려하였을 때, 거래지와 집까지의 거리가 멀다고, fraud가 더 많이 발생하는 것은 아닌 것으로 보인다.

```python
fraud_cnt_hour = (df.loc[df.is_fraud == 1]).trans_hour.value_counts()
total_cnt_hour = df.trans_hour.value_counts()

hour = total_cnt_hour.index
rate = []
for h in hour:
    fraud_rate = fraud_cnt_hour[h] / total_cnt_hour[h] * 100
    rate.append(fraud_rate)
```

### 추가 전처리 없는 base_line 제작

가장 기본적인 base_line을 바탕으로 하여, 해당 기준보다 좋은 모델을 선택하려고 한다.

```python
df_Xc = pd.concat([pd.get_dummies(df_X[['category', 'gender', 'gen', 'trans_hour']]), df_X[['amt']]], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_Xc, df_y, test_size = 0.2, stratify = df_y, random_state = 42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
```

```python
sum(pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(pred, y_test)
```

```python
# accuracy_score, precision_score, recall_score, f1_score
def get_acc(pred, test):
    acc = accuracy_score(pred, test)
    prec = precision_score(pred, test)
    rec = recall_score(pred, test)
    f1 = f1_score(pred, test)
    
    print('accuracy = %.3f' % acc)
    print('precision = %.3f' % prec)
    print('recall = %.3f' % rec)
    print('f1-score = %.3f' % f1)
```

```python
get_acc(pred, y_test)
```

base-line에서는 사기라고 말한 데이터의 79.7%가 정확했으며, 실제 사기 건수의 76.9%를 잡아 냈다.

### 이상치에 대한 세부 반영

다음과 같이 진행하려고 한다.
- category 별로 amt를 계산한다.  
- amt의 (Q3-Q1)의 1.5배의 값을 확인한다.  
- Q3 + 1.5배의 커트라인을 확인한다.
- 각 category별로 over했는지 확인하는 변수를 만든다.

이 때, test data에 대해서는 확인해서는 안되므로 train_set에 대해서만 해당 영향을 반영한다.

```python
df_Xc
```

```python
# 1) loc을 통해 1인 케이스만 분리
# 2) 이상치 계산
# 3) 이상치에 해당하는 줄의 넘버를 받는다.
# 4) 그 줄에만 1을 부여, 아닌 경우는 0을 부여한다.
cols = df_Xc.columns[:14]
for col in cols:
    num = df_Xc.loc[df_Xc[col] == 1].amt.values
    cut = ((np.quantile(num, .75) - np.quantile(num, .25))*1.5 + np.quantile(num, .75))
    new_col = col + '_over'
    
    get_overidx = list(df_Xc.loc[(df_Xc[col] == 1) & (df_Xc.amt > cut)].index)
    df_Xc[new_col] = [1 if i in get_overidx else 0 for i in range(1296675)]
```
